# if - elif - else statments


```python
age = int(input("Enter your age: "))

if age >= 18:                 # if is used to check condition
    print("you are adult.")

elif age >= 15 :              # elif is used to check multiple conditions
    print("you are minor.")  

else:                         # else is used to handle all other cases
 
    print("you are child.")      
```
